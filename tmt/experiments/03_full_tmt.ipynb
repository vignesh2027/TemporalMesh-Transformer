# Experiment 03 — Full TMT Training Run
All three innovations active: Mesh Attention + Temporal Decay + Adaptive Depth Routing.

```python
import torch
from tmt.model.config import TMTConfig
from tmt.training.trainer import TMTTrainer, TrainConfig
from tmt.data.dataset import load_text_dataset

cfg = TMTConfig(
    vocab_size=50258, d_model=512, n_heads=8, n_layers=12,
    graph_k=8, decay_rate=0.1, exit_threshold=0.85,
    dual_stream=True, memory_anchors=16, ffn_stream_dim=256,
)
print(cfg)
print(f'Device: {"cuda" if torch.cuda.is_available() else "cpu"}')
```

```python
loaders = load_text_dataset('wikitext-2', seq_len=256, batch_size=16)

train_cfg = TrainConfig(
    total_steps=10_000,
    warmup_steps=500,
    eval_every=500,
    save_every=1000,
    use_wandb=False,   # set True and login to wandb first
)

trainer = TMTTrainer(cfg, train_cfg, loaders['train'], loaders.get('validation'))
trainer.train()

tmt_ppl = trainer.evaluate()
print(f'Full TMT perplexity: {tmt_ppl:.2f}')
```
