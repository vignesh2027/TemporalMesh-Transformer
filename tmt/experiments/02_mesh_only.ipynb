# Experiment 02 — Ablation: Mesh Attention Only
TMT with only Innovation 1 (mesh attention) to isolate its contribution.

```python
import torch
from tmt.model.config import TMTConfig
from tmt.model.model import TMTModel
from tmt.training.trainer import TMTTrainer, TrainConfig
from tmt.data.dataset import load_text_dataset

# Disable innovations 2 and 3: set exit_threshold high, decay_rate=0
cfg = TMTConfig(
    vocab_size=50258, d_model=512, n_heads=8, n_layers=6,
    graph_k=8, decay_rate=0.0, exit_threshold=1.1,  # never exit early
    dual_stream=False, memory_anchors=0,
    ffn_stream_dim=256,
)
print(cfg)
```

```python
loaders = load_text_dataset('wikitext-2', seq_len=256, batch_size=16)
train_cfg = TrainConfig(total_steps=2000, warmup_steps=200, use_wandb=False)

trainer = TMTTrainer(cfg, train_cfg, loaders['train'], loaders.get('validation'))
trainer.train()
mesh_only_ppl = trainer.evaluate()
print(f'Mesh-only perplexity: {mesh_only_ppl:.2f}')
```
